## Step 2: Import Weather Data

The weather_prediction_dataset.csv from Zenodo describes various meteorological variables for daily records across 18 European cities from 2000 to 2010. This dataset is used to predict future weather conditions.

```python
import pandas as pd

# URL of Dataset on Zenodo, across locations in Europe
url = 'https://zenodo.org/records/4770937/files/weather_prediction_dataset.csv'

# Load Dataset
df = pd.read_csv(url)

# Display basic information about dataset
df.info()
```

```python
# Display Preview and Summary of Dataset

print_header("Dataset Preview")
display(df.head().style.set_caption("First 5 Rows").background_gradient(cmap='Blues'))

print_header("Dataset Summary")
display(df.describe().T.style.bar(color='#5fba7d'))
```

## Step 3: Check Missing Values

```python
print_header("Data Types & Missing Values")
display(pd.DataFrame({
    'Data Type': df.dtypes,
    'Missing Values': df.isna().sum(),
    '% Missing': (df.isna().mean()*100).round(2)
}))
```

## Step 4: Data Preprocessing

```python
# 1. Handle Missing Values
df = df.dropna() # Drop rows with missing values
```

```python
# 2. Encode Categorical Data (One-Hot Encoding Method) - Turn categorical data into numerical data
df = pd.get_dummies(df, drop_first=True) # Drop first column
```

```python
# 3. Visualize Correlation Matrix
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot=False)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```

**Correlation Matrix**

A correlation matrix is a table showing correlation coefficients (r) between pairs of variables. Each value ranges from -1 to 1 and tells you how strongly and in what direction two variables are related.

**Correlation Value (r) Interpretation**

+1.0 > Perfect positive correlation <br>
+0.7 to +0.9 > Strong positive correlation <br>
+0.4 to +0.6 > Moderate positive correlation <br>
+0.1 to +0.3 > Weak positive correlation <br>
0 > No correlation <br>
-0.1 to -0.3 > Weak negative correlation <br>
-0.4 to -0.6 > Moderate negative correlation <br>
-0.7 to -0.9 > Strong negative correlation <br>
-1.0 > Perfect negative correlation <br>


```python
# 4. Feature Selection
# Define Features and Target Variable

X1 = df.drop('BASEL_temp_mean', axis=1) #axis is column
y1 = df['BASEL_temp_mean']

X2 = df.drop('BASEL_humidity', axis=1)
y2 = df['BASEL_humidity']

X3 = df.drop('BASEL_pressure', axis=1)
y3 = df['BASEL_pressure']
```

## Step 5: Split the Data

```python
from sklearn.model_selection import train_test_split

# Split Data Temperature
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=42)

# Split Data Humidity
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)

# Split Data Pressure
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size=0.2, random_state=42)
```

## Step 6: Build and Train the Model

```python
from sklearn.linear_model import LinearRegression

# Init Temperature Model
model_temp = LinearRegression()
# Train Temperature Model
model_temp.fit(X1_train, y1_train)

# Init Humidity Model
model_humidity = LinearRegression()
# Train Humidity Model
model_humidity.fit(X2_train, y2_train)

# Init Pressure Model
model_pressure = LinearRegression()
# Train Pressure Model
model_pressure.fit(X3_train, y3_train)
```

## Step 7: Evaluate the Model

```python
from sklearn.metrics import mean_squared_error, r2_score

# -----Temperature Predictions-----
y1_pred = model_temp.predict(X1_test)

# Calc Mean Squared Error
mse_1 = mean_squared_error(y1_test, y1_pred)

# Calc R-Squared Score
r2_1 = r2_score(y1_test, y1_pred)

# Print Results
print("Temperature Evaluation")
print(f"Mean Squared Error: {mse_1}")
print(f"R2 Score: {r2_1}")
print()

# -----Humidity Predictions-----
y2_pred = model_humidity.predict(X2_test)

# Calc Mean Squared Error
mse_2 = mean_squared_error(y2_test, y2_pred)

# Calc R-Squared Score
r2_2 = r2_score(y2_test, y2_pred)

# Print Results
print("Humidity Evaluation")
print(f"Mean Squared Error: {mse_2}")
print(f"R2 Score: {r2_2}")
print()

# -----Pressure Predictions-----
y3_pred = model_pressure.predict(X3_test)

# Calc Mean Squared Error
mse_3 = mean_squared_error(y3_test, y3_pred)

# Calc R-Squared Score
r2_3 = r2_score(y3_test, y3_pred)

# Print Results
print("Pressure Evaluation")
print(f"Mean Squared Error: {mse_3}")
print(f"R2 Score: {r2_3}")
print()
```

**Term Definition**

Mean Square Error is measuring how far away the predicted value is from the actual value. The lower the value, the better the model <br>
R2 Score is a measure of how well the model fits the data <br>

## Step 8: Visualise the Results

```python
import matplotlib.pyplot as plt

def plot_regression(y_true, y_pred, title, unit):
    plt.figure(figsize=(6, 6)) # Set figure size
    plt.scatter(y_true, y_pred, color='skyblue', edgecolors='k', alpha=0.7)
    max_val = max(max(y_true), max(y_pred))
    min_val = min(min(y_true), min(y_pred))
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Ideal Prediction')

    plt.xlabel(f'Actual Values ({unit})')
    plt.ylabel(f'Predicted Values ({unit})')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

# Temperature Plot (C)
plot_regression(y1_test, y1_pred, 'Actual Temperature vs Predicted Temperature', 'C')

# Humidity Plot (%)
plot_regression(y2_test, y2_pred, 'Actual Humidity vs Predicted Humidity', '%')

# Pressure Plot (hPa)
plot_regression(y3_test, y3_pred, 'Actual Pressure vs Predicted Pressure', 'hPa')
```
